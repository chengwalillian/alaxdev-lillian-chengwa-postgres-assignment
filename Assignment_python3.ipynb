```python
## 16.Reverse a list (without using the built-in reverse method).

def reverse_list(input_list):
    reversed_l = []
    for i in range(len(input_list) - 1, -1, -1):
        reversed_l.append(input_list[i])
    return reversed_l

numbers = [1, 2, 3, 4]
print(reverse_list(numbers))
```

```python
##  36 Ask user to input three numbers and store them in a list. Then, convert the list to a set and check if any duplicates were removed; tell the user if there were duplicates.

numbers = []
for i in range(3):
    while True:
        try:
            num = int(input(f"Enter number {i+1}: "))
            numbers.append(num)
            break
        except ValueError:
            print("Invalid input. Please enter a whole number.")

unique_numbers = set(numbers)
if len(unique_numbers) < len(numbers):
    print("There were duplicates in your input.")
else:
    print("No duplicates were found.")

print("Your unique numbers are:", unique_numbers)
```

```python
#39 Given a list of numbers, ask the user to input the numbers, then compute: number of positive, negative, and zero values using conditionals.



numbers = input("Enter numbers separated by spaces: ")
numbers = list(map(int, numbers.split()))

positive = 0
negative = 0
zeros = 0

for num in numbers:
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1
    else:
        zeros += 1

print("Positive numbers:", positive)
print("Negative numbers:", negative)
print("Zeros:", zeros)

```
